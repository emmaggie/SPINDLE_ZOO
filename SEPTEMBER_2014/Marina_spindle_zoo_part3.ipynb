**For radar plot we need (1) means of numeric columns. We need to (2) subtract the _CAT columns but (3) put back the organism column for grouping.**

```python
#these are the potential numeric variables; REMOVE CATs
dfo._get_numeric_data().columns
```

```python
#get all columns that end with CAT (or their position in the data frame)
pattern=re.compile(r"(.+_CAT)")
#print pattern.match('organism_CAT').group()
```

```python
cols_to_drop=[]
for column in dfo._get_numeric_data().columns:
    #print column
    if pattern.match(column) != None:
        cols_to_drop.append(pattern.match(column).group())
```

```python
cols_to_drop
```

```python
#sanity check
dfo._get_numeric_data()[cols_to_drop].head()
```

```python
dfo_for_pygal=dfo._get_numeric_data().drop(cols_to_drop,axis=1)
```

```python
dfo.info()
```

```python
dfo_for_pygal.info()
```

```python
dfo_for_pygal.head(2)
```

```python
dfo_for_pygal['organism']=dfo.organism
#dfo.organism
```

```python
dfo_for_pygal_2=dfo_for_pygal.drop('num_of_cells_NUM',axis=1)
```

```python
dfo_for_pygal_2.info()
```

```python
grouped_org_mei=dfo_for_pygal_2.groupby(['organism','meiotic'],sort=False)
```

```python
len(grouped_org_mei)
```

```python
#this is a data frame with the means per organism, split by meiotic/mitotic 
means=grouped_org_mei.mean()
```

```python
#test the plot on one organism
log_means_1=[np.log10(x) for x in means.loc['xenopus_laevis',1].values]
log_means_0=[np.log10(x) for x in means.loc['xenopus_laevis',0].values]
```

```python
means.loc['xenopus_laevis',1].values
```

```python
len(list(means.columns)), len(log_means_0)
```

```python
#try pygal
##http://pygal.org/builtin_styles/
#https://github.com/Kozea/pygal/blob/master/pygal/config.py

radarChart1=pygal.Radar()
radarChart1.config.x_labels=list(means.columns)
radarChart1.config.width=800
radarChart1.config.height=600
radarChart1.add('meiotic',log_means_1)
radarChart1.add('mitotic',log_means_0)
#http://code.activestate.com/recipes/578123-draw-svg-images-in-python-python-recipe-enhanced-v/
SVG(radarChart1.render(), height=300, width=400) #drop time_pf, chromosomes,genome

#radarChart1.x_labels
```
