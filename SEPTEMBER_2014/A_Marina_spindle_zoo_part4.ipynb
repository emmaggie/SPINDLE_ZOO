```python
#get all columns that end with CAT (or their position in the data frame)
pattern=re.compile(r"(.+_CAT)")
#print pattern.match('organism_CAT').group()
```

```python
cols_to_drop=[]
for column in dfo_mit._get_numeric_data().columns:
    #print column
    if pattern.match(column) != None:
        cols_to_drop.append(pattern.match(column).group())
```

```python
dfo_for_pygal=dfo_mit._get_numeric_data().drop(cols_to_drop,axis=1)
```

```python
dfo_for_pygal.num_of_cells_NUM.hist(bins=20) #drop this one too?
```

```python
dfo_for_pygal_2=dfo_for_pygal.drop('num_of_cells_NUM',axis=1)
```

```python
dfo_for_pygal_2.columns
```

```python
dfo_for_pygal_2['organism']=dfo_mit.organism
```

```python
dfo_for_pygal_2['stage']=dfo_mit.stage
```

```python
grouped_org_mit=dfo_for_pygal_2.groupby(['organism','stage'],sort=False)
```
