###Load data

###Map colors

```python
%%R
gg_color_hue <- function(n) {
    #Generates uniformly ditributed colors on the hues color wheel
    #This is the color assignment used by ggplot2

    hues = seq(15, 375, length=n+1)
    hcl(h=hues, l=65, c=100)[1:n]

}
```

```python
%%R
original$organism<-as.character(original$organism)
cols=gg_color_hue(length(unique(original$organism)))
```

```python
%%R
print(cols)
plot(1:length(unique(original$organism_CAT)), pch=16, cex=2, col=cols)
my_labels=unique(original$organism)
```

```python
%%R
mapping_vector_lab<-setNames(object=cols,nm=my_labels)
```

```python
%%R
plot(1:length(mapping_vector_lab), pch=16, cex=2, col=cols)
text(1:length(mapping_vector_lab),names(mapping_vector_lab),cex=0.5) 
```

```python
%%R
lookup_close_colors<-function(orig_mapping_vector,string_color_1, string_color_2){
    val_1=which(orig_mapping_vector==string_color_1)
    val_2=which(orig_mapping_vector==string_color_2)
    if ( (val_1==val_2+1) | (val_1==val_2-1) ){
        return('change shape!')
        }
    else{
        return('no change of shape required')    
        }
}
```
