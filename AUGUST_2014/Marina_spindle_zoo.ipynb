###SUMMARIZE THE DATA

```python
#df3 is df2 with 'na' replaced with NaN
df3.describe()
#this is a description of all the numerical data
#let's find which one has the highest SD relative to mean.
#at this staeg we ingnore teh fact that we have several species, the variable with highest might be the one of interest
```

```python
df3.describe().shape
#14 variables are numerical. The values above do not assume substitution with median
#Let's have a look at the %SD
```

```python
print df3.describe().ix['std'][0:5], type(df3.describe().ix['std']) #object of type series
print df3.describe().ix['mean'][0:5], type(df3.describe().ix['mean']) #object of type series
```

```python
#below you can see SD as % of the mean. Idea is that the higher SD, the higher probability that the feature matters.
print ((df3.describe().ix['std']/df3.describe().ix['mean'])*100).order(ascending=False)[0:6]
#print type(((df3.describe().ix['std']/df3.describe().ix['mean'])*100).order(ascending=False)[0:6])
```

```python
#Let's load all of these to variables and then, the data frame:
#df3.describe()
d3_sum=df3.describe()
std_overall=d3_sum.ix['std']
mean_overall=d3_sum.ix['mean']

#print std_overall, type(std_overall)
#print mean_overall, type(mean_overall)
```

```python
#Combining the two into a series
##http://pandas.pydata.org/pandas-docs/stable/merging.html
#http://stackoverflow.com/questions/18062135/combining-two-series-into-a-dataframe-in-pandas
#http://stackoverflow.com/questions/16945166/merging-two-time-series-in-pandas
#http://en.wikipedia.org/wiki/Join_(SQL)
#http://www.w3schools.com/sql/sql_join_inner.asp
mean_std_overall=pd.concat([std_overall,mean_overall],axis=1)
mean_std_overall['Relative_STD']=mean_std_overall['std']/mean_std_overall['mean']*100 
```

```python
print mean_std_overall.sort_index()
```
